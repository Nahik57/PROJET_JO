```python
## Importation des bibliothèques ##

import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd
import plotly.express as px
import seaborn as sns
```

```python
## Importation du Dataset ##

joml = pd.read_csv("athlete_events.csv")
```

```python
## Visualisation du Dataset ##

joml
```

```python
## Visualisation des différentes informations de l'ensemble de données, 
## comme le type de données (numérique, catégorielle) : colonne "Dtype",
## comme les valeurs manquantes : colonne "Non-Null Count",
## etc.... ##

joml.info()
```

```python
## Suppression des lignes contenant "Winter" dans la colonne "Season" ##

joml = joml[joml['Season'] != 'Winter']
joml
```

```python
## Suppression de la colonne "Games" qui fait doublon avec les infos des colonnes "Year" et "Season" ##

joml = joml.drop("Games", axis=1)
joml
```

```python
joml = joml.dropna(subset=['Age', 'Height', 'Weight'])
joml
```

```python
## Visualisation des différentes informations de l'ensemble de données, 
## comme le type de données (numérique, catégorielle) : colonne "Dtype",
## comme les valeurs manquantes : colonne "Non-Null Count",
## etc.... ##

joml.info()
```

```python
## Suppression des sports collectifs ou trop ancien ##

spcol = ['Art Competitions', 'Baseball', 'Basketball', 'Beach Volleyball', 'Football', 'Handball', 'Hockey', 'Ice Hockey', 'Lacrosse', 'Rowing',
        'Rugby', 'Rugby Sevens', 'Softball', 'Synchronized Swimming', 'Tug-Of-War', 'Volleyball', 'Water Polo']

joml2 = joml[~joml['Sport'].isin(spcol)]

joml2
```

```python
## Ajout du sport Rowing pour les épreuves 'single' ##

rowing_single = joml[(joml["Sport"] == "Rowing") & (joml["Event"].str.contains("single", case=False))]
joml = pd.concat([rowing_single, joml2])
joml
```

```python
## Suppression des épreuves collectives ##

epcol = ['Synchronized', 'Doubles', 'Fours', 'Team', 'Relay', 'Group', 'Pairs', 'Quadruple', 'Three', 'Two', 'Mixed', 'Skiff', 'Tandem']

joml = joml[joml['Event'].apply(lambda x: not any(term in x for term in epcol))]

joml
```

```python
## Suppression des 1ere infos de la colonne "Event" jusqu'à " 's " ##

joml.loc[:, 'Event'] = joml['Event'].str.split("'s", n=1).str[1].str.strip()
joml
```

```python
## Arrondi à l'unité des colonnes 'Age' et 'height" ##

colonnes_arrondis = ['Age', 'Height']
for colonne in colonnes_arrondis:
    joml[colonne] = joml[colonne].astype(str).str.split('.').str[0].astype(int)
joml
```

```python
## tri du dataset par la colonne 'ID' ##

joml = joml.sort_values(by='ID', ascending = True)
joml
```

```python
## Suppression des lignes avant 1952 ##

joml = joml[joml['Year'] >= 1952]
joml
```

```python
## calcul de l'IMC par ligne et ajout du résultat dans une nouvelle colonne ##

joml['IMC'] = joml['Weight'] / (joml['Height'] / 100) ** 2
joml
```

```python
## Limiter à deux décimales la colonne "IMC"  ##

joml['IMC'] = joml['IMC'].round(2)
joml
```

```python
## déplacer la colonne IMC à coté de la colonne Weight ##

col_to_move = 'IMC'
new_index = 6
cols = list(joml.columns)
cols.insert(new_index, cols.pop(cols.index(col_to_move)))
joml = joml[cols]
joml
```

```python
## Visualisation du nombre de lignes par Event ##

event_counts = joml['Event'].value_counts()
print(event_counts)
```

```python
## Suppression des Event ayant 10 linges ou moins ##

events_to_keep = event_counts[event_counts >= 10].index
joml = joml[joml['Event'].isin(events_to_keep)]
joml
```

```python
## Visualisation du nombre de lignes par Event ##

event_counts = joml['Event'].value_counts()
print(event_counts)
```

```python
# Remplacer 'Bronze', 'Silver', 'Gold' par 'Yes' dans la colonne "Medal" ##

joml['Medal'] = joml['Medal'].replace({'Bronze': 'Yes', 'Silver': 'Yes', 'Gold': 'Yes'})
joml
```

```python
## Remplacement les "NaN" par "No" dans la colonne "Medal" ##

joml['Medal'] = joml['Medal'].fillna('No')
joml
```

```python
## ajouter une colonne indiquant si les JO se déroulent dans le pays de l'athlète 1/4 ##

## Charger le fichier CSV issu du site "SimpleMaps" ##

cities_df = pd.read_csv('worldcities.csv')
cities_df.head()
```

```python
## ajouter une colonne indiquant si les JO se déroulent dans le pays de l'athlète 2/4 ##

## Créer un dictionnaire avec les villes comme clés et les codes pays comme valeurs ##

city_country_dict = dict(zip(cities_df['city'], cities_df['iso3']))
city_country_dict
```

```python
## ajouter une colonne indiquant si les JO se déroulent dans le pays de l'athlète 3/4 ##

def check_city_in_country(city, country_code):
    return city_country_dict.get(city) == country_code

## Appliquer cette fonction à chaque ligne du dataframe ##

joml['City_in_Country'] = joml.apply(lambda row: check_city_in_country(row['City'], row['NOC']), axis=1)
joml
```

```python
## ajouter une colonne indiquant si les JO se déroulent dans le pays de l'athlète 4/4 ##

## déplacer la colonne 'City_in_Country' à coté de la colonne "City" ##

col_to_move = 'City_in_Country'
new_index = 12
cols = list(joml.columns)
cols.insert(new_index, cols.pop(cols.index(col_to_move)))
joml = joml[cols]
joml
```

```python
## Recherche et visualisation des éventuels doublons ##

duplicateRows = joml[joml.duplicated()]
duplicateRows
```
